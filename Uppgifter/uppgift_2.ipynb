```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../Data/athlete_events.csv")

selected_sports = ["Basketball", "Swimming", "Athletics", "Gymnastics"]


filtered_sports = df[df["Sport"].isin(selected_sports) & df["Medal"].notnull()]

medals_by_country_sport = (
    filtered_sports.groupby(["Sport", "NOC"]).size().reset_index(name="Count")
)


fig = px.bar(
    medals_by_country_sport,
    x="NOC",
    y="Count",
    color="Sport",
    title="Medaljfördelning mellan länder i valda sporter",
    labels={"NOC": "Land", "Count": "Antal medaljer", "Sport": "Sport"},
)

fig.show()
```

```python

filtered_ages = df[df["Sport"].isin(selected_sports) & df["Age"].notnull()]


fig = px.histogram(
    filtered_ages,
    x="Age",
    color="Sport",
    nbins=20,
    title="Åldersfördelning i valda sporter",
    labels={"Age": "Ålder", "count": "Antal idrottare", "Sport": "Sport"},
)

fig.show()
```

```python

participants_per_sport_os = (
    df[df["Sport"].isin(selected_sports)]
    .groupby(["Year", "Sport"])
    .size()
    .reset_index(name="Count")
)


fig = px.line(
    participants_per_sport_os,
    x="Year",
    y="Count",
    color="Sport",
    title="Antal deltagare per OS för varje sport",
    labels={"Year": "År", "Count": "Antal deltagare", "Sport": "Sport"},
)

fig.show()
```

```python

medals_by_gender_sport = (
    filtered_sports.groupby(["Sport", "Sex"]).size().reset_index(name="Count")
)


fig = px.bar(
    medals_by_gender_sport,
    x="Sport",
    y="Count",
    color="Sex",
    title="Medaljer per kön i valda sporter",
    labels={"Sport": "Sport", "Count": "Antal medaljer", "Sex": "Kön"},
    barmode="group",
)

fig.show()
```
